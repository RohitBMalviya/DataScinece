```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
irisDf = pd.read_csv('iris.csv')
```

```python
irisDf.drop(columns='Id',inplace=True)
```

```python
# Part-I
print(irisDf.dtypes)
```

```python
# Part_II
irisDf.hist()
plt.suptitle('Histograms of Iris Dataset Features')
plt.tight_layout()
plt.show()
```

```python
# Part-III
irisDf.boxplot()
plt.title('Boxplots of Iris Datasets Features')
plt.ylabel('Centimeters')
plt.show()
```

```python
# Part-IV
plt.scatter(irisDf['SepalLengthCm'][irisDf['Species']=='Iris-setosa'], irisDf['SepalWidthCm'][irisDf['Species']=='Iris-setosa'], color='red', label='Iris-setosa')
plt.scatter(irisDf['SepalLengthCm'][irisDf['Species']=='Iris-versicolor'], irisDf['SepalWidthCm'][irisDf['Species']=='Iris-versicolor'], color='blue', label='Iris-versicolor')
plt.scatter(irisDf['SepalLengthCm'][irisDf['Species']=='Iris-virginica'], irisDf['SepalWidthCm'][irisDf['Species']=='Iris-virginica'], color='green', label='Iris-virginica')
plt.legend()
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()
plt.scatter(irisDf['PetalLengthCm'][irisDf['Species']=='Iris-setosa'], irisDf['PetalWidthCm'][irisDf['Species']=='Iris-setosa'], color='red', label='Iris-setosa')
plt.scatter(irisDf['PetalLengthCm'][irisDf['Species']=='Iris-versicolor'], irisDf['PetalWidthCm'][irisDf['Species']=='Iris-versicolor'], color='blue', label='Iris-versicolor')
plt.scatter(irisDf['PetalLengthCm'][irisDf['Species']=='Iris-virginica'], irisDf['PetalWidthCm'][irisDf['Species']=='Iris-virginica'], color='green', label='Iris-virginica')
plt.legend()
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```
