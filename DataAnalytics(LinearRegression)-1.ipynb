```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
boston = pd.read_csv('boston_housing.csv')
```

```python
# sns.pairplot(boston, vars = ['rm', 'zn', 'black', 'dis', 'chas','medv'])
```

```python
X = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'medv', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']]
Y = boston['age']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.5)
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
```

```python
prediction = lm.predict(X_test)
plt.scatter(Y_test,prediction)
plt.xlabel('Y Test')
plt.xlabel('Predicted Y')
```

```python
print('Mean Absolute Error: ', metrics.mean_absolute_error(Y_test,prediction))
print('Mean Squared Error: ', metrics.mean_squared_error(Y_test, prediction))
print('Mean Squared root Error: ', np.sqrt(metrics.mean_squared_error(Y_test,prediction)))
```

```python
sns.histplot((Y_test-prediction),bins=50);
```
