```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanicDf = pd.read_csv('titanic.csv')
```

```python
group = titanicDf.groupby('Sex').describe()
print(group[['Survived','Age']])
```

```python
sns.boxplot(data=titanicDf, x='Sex',y='Age',hue='Survived')
plt.title('Age Distribution by Gender & Survived')
plt.show()
```
