#  Number of persons statewise vaccinated for first dose in India

```python
firstdose = df.groupby('State')['First Dose Administered'].sum()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(firstdose.index, firstdose.values, label="First Dose")
ax.set_xlabel("State")
ax.set_ylabel("Number of First Dose Administered")
ax.set_title("Number of Individuals Statewise Vaccinated for First Dose")

ax.legend()
plt.xticks(rotation=90)
plt.show()
```

#  Number of persons statewise vaccinated for second dose in India

```python
Seconddose = df.groupby('State')['Second Dose Administered'].sum()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(Seconddose.index, Seconddose.values, label="Second Dose")
ax.set_xlabel("State")
ax.set_ylabel("Number of Second Dose Administered")
ax.set_title("Number of Individuals Statewise Vaccinated for second Dose")

ax.legend()
plt.xticks(rotation=90)
plt.show()
```

# Number of Males vaccinated

```python
malevacinated = newdf['Male (Doses Administered)'].sum()
malevacinated
```

```python
fig, ax = plt.subplots()

ax.bar(['Males'], [malevacinated])

ax.set_xlabel("Gender")
ax.set_ylabel("Number of Individuals Vaccinated")
ax.set_title("Number of Individuals Vaccinated by Male")
```

# Number of Female vaccinated

```python
femalevacinated = newdf['Female (Doses Administered)'].sum()
femalevacinated
```

```python
fig, ax = plt.subplots()

ax.bar(['Female'], [femalevacinated])

ax.set_xlabel("Gender")
ax.set_ylabel("Number of Individuals Vaccinated")
ax.set_title("Number of Individuals Vaccinated by female")
```

# Comparsion between male and female vaccinated

```python
fig, ax = plt.subplots()
ax.bar(["Males", "Females"], [malevacinated, femalevacinated])
ax.set_xlabel("Gender")
ax.set_ylabel("Number of Individuals Vaccinated")
ax.set_title("Number of Individuals Vaccinated by Gender")
```
