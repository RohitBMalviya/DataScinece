```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanicDf = pd.read_csv('titanic.csv')
```

```python
# Part-I
sns.heatmap(titanicDf.corr(),annot=True,cmap='Blues')
```

```python
titanicDf.isnull().sum()
```

```python
titanicDf.describe(include=['O'])
```

```python
titanicDf[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean()
```

```python
titanicDf[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean()
```

```python
titanicDf[['SibSp', 'Survived']].groupby(['SibSp'], as_index=False).mean()
```

```python
titanicDf[['Parch', 'Survived']].groupby(['Parch'], as_index=False).mean()
```

```python
grid = sns.FacetGrid(titanicDf, col='Survived', row='Pclass')
grid.map(plt.hist, 'Age', bins=20)
grid.add_legend();
```

```python
# Part-II
sns.histplot(titanicDf['Fare'],kde=False)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Distribution of Fare')
plt.show()
```
